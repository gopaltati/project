```python
# Full Workflow
missing_columns = df.isnull().sum()[df.isnull().sum() > 0]
numeric_columns =df.select_dtypes(include='number').columns.tolist()
categorical_columns = df.select_dtypes(include='object').columns.tolist()
duplicate_columns = df.columns[df.T.duplicated()].tolist()

print("Columns with Missing Values:")
print(missing_columns)
print("\nNumeric Columns:")
print(numeric_columns)
print("\nCategorical Columns:")
print(categorical_columns)
print("\nDuplicate Columns:")
print(duplicate_columns)

# Clean dataset
data_cleaned = df.loc[:, ~df.T.duplicated()]
print("\nData After Removing Duplicate Columns:")
print(data_cleaned)
```

```python
# Identify duplicate columns
duplicate_columns = df.columns[df.T.duplicated()].tolist()
print("Duplicate Columns Before Removal:", duplicate_columns)

# Remove duplicate columns
data_no_duplicates = df.loc[:, ~df.T.duplicated()]
print("Data Shape Before Removing Duplicates:", df.shape[]
print("Data Shape After Removing Duplicates:", data_no_duplicates.shape)
```

```python
# Identify constant columns
constant_columns = [col for col in df.columns if df[col].nunique() == 1]
print("Constant Columns Before Removal:", constant_columns)

# Remove constant columns
df_no_constants = df.drop(columns=constant_columns)
print("Data Shape Before Removing Constants:", df.shape)
print("Data Shape After Removing Constants:", df_no_constants.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Numeric columns
numeric_columns = df.select_dtypes(include='number').columns.tolist()

# Create box plots
for col in numeric_columns:
    plt.figure()
    sns.boxplot(df[col])
    plt.title(f"Box Plot for {col}")
    plt.show()
```

```python
# Select 6 columns (can be adjusted based on your dataset)
selected_columns = numeric_columns[:3] + categorical_columns[:3]

for col in selected_columns:
    plt.figure()
    if df[col].dtype == 'object':
        sns.countplot(x=df[col], order=df[col].value_counts().index)
    else:
        sns.histplot(df[col], kde=True)
    plt.title(f"Distribution for {col}")
    plt.show()
```
